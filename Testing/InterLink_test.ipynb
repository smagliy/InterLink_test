```python
import pandas as pd
df = pd.read_csv("acme_worksheet.csv")
        
df.head(100)
```

```python
pvt = pd.pivot_table(df, values = 'Work Hours', index=['Employee Name'], columns = 'Date').reset_index()
pvt.head(100)

```

```python
pvt=pvt.rename({'Employee Name': 'Name/Date', 'Jul 01 2020': '2020-07-01', 'Jul 02 2020': '2020-07-02', 'Jul 03 2020': '2020-07-03', 'Jul 04 2020': '2020-07-04', 'Jul 05 2020': '2020-07-05', 'Jun 29 2020': '2020-06-29', 'Jun 30 2020': '2020-06-30'}, axis=1)
pvt.head()
```

```python
pvt.to_csv("finished file.csv")
```
